```python
import pandas as pd # pandas is a package and is helpful for data manipulation
import numpy as np
import matplotlib.pyplot as plt

hospital_file = "./HospitalAdmissionsData.csv";
hospital = pd.read_csv(hospital_file)
```

```python
#PART A
hospital.columns
```

```python
#PART B and C
#There is one float and 3 ints
#The object features are Admission_Type, Insurance_Type, Religion_Type, Race, Dx
hospital.info()
```

```python
#PART D
#The types are private, medicare, goverment, medicaid, and self pay
hospital["Insurance_Type"].unique()
```

```python
#PART E
#Here is the info
hospital["AdmissionLengthDays"].describe()
```

```python
#PART F
#The most common are emergency, medicare, catholic, white, and newborn
hospital.describe(include='object')
```

```python
#PART G 
#Without log scale
#ALmost all admissions are between 0-25 days, after that the count significantly drops off
hospital.hist(column='AdmissionLengthDays',sharex=True, label=True)
plt.title('Histogram of Admission Length Days')
plt.xlabel('Admission Length Days')
```

```python
#PART G
#With log Scale
hospital.hist(column='AdmissionLengthDays',sharex=True, label=True)
plt.title('Histogram of Admission Length Days')
plt.xlabel('Admission Length Days')
plt.xscale('log')
```

```python
#PART H
#Whether they died or not doesn't seem to effect the length of admission
hospitalbar = hospital.groupby('Death_1').mean()
hospitalbar.plot.bar(y=['AdmissionLengthDays'])
plt.title('Average Admission Length Days by Death')
plt.ylabel('Average Admission Length Days')
plt.xlabel('Died or not (0 = Survived)')
```
